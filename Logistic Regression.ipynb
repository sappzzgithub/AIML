```python
import pandas as pd
```

```python
df=pd.read_csv("/content/heart.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#Logistic Regression and train_test_split:
#When performing Logistic Regression, the dataset needs to be split into two sets:
#Training Set: Used to train the Logistic Regression model (fit the model to the data).
#Testing Set: Used to evaluate how well the trained model generalizes to unseen data.



from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 0:-1], df[df.columns[-1]], test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression   #importing logisticRegession
from sklearn.tree import DecisionTreeClassifier       #importing DecisionTreeClassifier

clf1 = LogisticRegression()   #initializing the logistic classifier, which will be used later for training and predicting.
clf2 = DecisionTreeClassifier()  #initializing Decision tree classifier,which will be used later for training and predicting.
```

```python
clf1.fit(x_train, y_train)   #this line trains the Logistic Regression model(clf1)using the features (x_train) and the target labels (y_train).

clf2.fit(x_train, y_train)   #this line trains the Decision Tree model (clf2) using the same training data (x_train, y_train).
```

```python
y_pred1 = clf1.predict(x_test)
y_pred2 = clf2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression:",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Tree:",accuracy_score(y_test,y_pred2))
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
```
