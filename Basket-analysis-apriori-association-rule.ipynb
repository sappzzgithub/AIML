```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules

```

```python
df=pd.read_csv("/content/bread basket.csv")
```

```python
df.head(10)
```

```python
df.tail()
```

```python
grouped_df=df.groupby(['Transaction','Item'])['Item'].count().reset_index(name='count')
basket_df=grouped_df.pivot_table(index='Transaction',columns='Item',values='count',aggfunc='sum').fillna(0)
basket_df.head()
```

```python
#defining the hot encoding function to make the data suitable for the concerned libraries
def hot_encode(x):
  if(x<=0):
    return 0
  if(x>=1):
    return 1
```

```python
#encode the dataset
basket_encoded=basket_df.applymap(hot_encode)
basket_df=basket_encoded
basket_df
```

```python
#apply data mining
```

```python
#building the model with min support=0.01(1%)
frq_items=apriori(basket_df,min_support=0.01,use_colnames=True)
frq_items
```

```python
#colleting the inferred rules in a dataframe
!pip install mlxtend
from mlxtend.frequent_patterns import association_rules
num_itemsets_dict={'itemsets':[2]}
rules=association_rules(frq_items,metric='confidence',min_threshold=0.01, num_itemsets=num_itemsets_dict)
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
rules.reset_index(drop=True,inplace=True)
rules


```
